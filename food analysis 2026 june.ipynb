```python
import requests
```

```python
url = "https://world.openfoodfacts.org/api/v2/search?countries_tags_en=germany&page_size=50"
```

```python
headers = {
   "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"   # creating a headers dictionary
}
```

```python
response = requests.get(url, headers=headers)# sending a request to get the data for this url and storing it in the response block
```

```python
type(response)
```

```python
data = response.json() # take the json data from API response and convert it into a dictory or list and store it in data variable
```

```python
data.keys() # show all the keys in the dictory
```

```python
# Check how many products are left (will be 49 now)
print("Products remaining:", len(data['products']))

# Get the FIRST product (index 0) safely
first_product = data['products'][0]
print("First product name:", first_product.get('product_name'))

# Look at ALL the keys of the first product
print("Keys in first product:", first_product.keys())
```

```python
# 1. Confirm we have 50 again
print("Count:", len(data['products']))

# 2. Import pandas (the "Excel on steroids" library)
import pandas as pd

# 3. Turn the list of dictionaries into a beautiful table
df = pd.DataFrame(data['products'])

# 4. Show the shape (rows, columns)
print("Shape of DataFrame:", df.shape)

# 5. Look at the first 5 rows (so you can see the actual data!)
df.head()
```

```python
# Show only the columns we need
cols_to_check = ['product_name', 'countries_tags', 'nutriscore_grade', 'packaging_tags', 'labels_tags']
df[cols_to_check].head(10)
```

```python
# Example for a single list
my_list = ['en:canada', 'en:france', 'en:germany']
print('en:germany' in my_list)  # True
```

```python
# Get all packaging tags, drop the empty ones
all_packaging = df['packaging_tags'].dropna()

# Create an empty list to store all materials
all_materials = []

# Loop through each product's packaging list
for tags in all_packaging:
    for tag in tags:
        # Remove the 'en:' prefix to get the clean material name
        clean_tag = tag.replace('en:', '')
        all_materials.append(clean_tag)

# Get unique materials and sort them
unique_materials = sorted(set(all_materials))
print("Unique packaging materials in this dataset:")
print(unique_materials)
```

```python
df = df[df['nutriscore_grade'] != 'not-applicable	']
df = df[df['countries_tags'].apply(lambda x: 'en:germany' in x)]

eco_keywords = ['glass', 'paper', 'cardboard', 'recyclable-metals', 'paperboard']

df['eco_packaging'] = df['packaging_tags'].apply(
    lambda x: 1 if any(keyword in str(x).lower() for keyword in eco_keywords) else 0
)
```

```python
#I defined eco‑packaging as materials that are glass, paper, or explicitly marked as recyclable metals. 
#I chose this strict definition to avoid labeling plastics or mixed materials as 'eco,' which would dilute the integrity of the analysis.
```

```python
print("Number of eco-friendly products:", df['eco_packaging'].sum())
print("\nLet's look at them:")
print(df[df['eco_packaging'] == 1][['product_name', 'packaging_tags']])
```

```python
# Convert letters to numbers
grade_map = {'a': 5, 'b': 4, 'c': 3, 'd': 2, 'e': 1}
df['nutriscore_num'] = df['nutriscore_grade'].map(grade_map)

# Check the average health score for eco vs non-eco
print(df.groupby('eco_packaging')['nutriscore_num'].mean())
```

```python
#For this analysis, 'eco-friendly packaging' is defined as packaging that includes glass, paper, cardboard, or explicitly recyclable metals. 
#We chose this definition because even if a product uses a plastic film, the presence of recyclable or biodegradable materials shows a genuine effort 
#toward sustainability.
```

```python
# List of countries to pull
countries = [ 'france', 'netherlands']

# Create an empty list to store dataframes
all_dfs = []

for country in countries:
    print(f"Pulling data for {country}...")
    url = f"https://world.openfoodfacts.org/api/v2/search?countries_tags_en={country}&page_size=50"
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        products = data['products']
        
        # Create a temporary dataframe
        temp_df = pd.DataFrame(products)
        
        # Add a column for country (so we know where each product came from)
        temp_df['country'] = country
        
        all_dfs.append(temp_df)
        print(f"  Got {len(products)} products from {country}")
    else:
        print(f"  Error fetching {country}: {response.status_code}")

# Combine all dataframes into one
df_all = pd.concat(all_dfs, ignore_index=True)
print(f"\nTotal products collected: {len(df_all)}")
```

```python
# 1. Add a country column to the original German data (since we didn't do that before)
df['country'] = 'germany'

# 2. Combine all three countries into one master DataFrame
df_master = pd.concat([df, df_all], ignore_index=True)

# 3. Check the total and distribution
print(f"Total products in master dataset: {len(df_master)}")
print("\nProducts per country:")
print(df_master['country'].value_counts())
```

```python
# Step 1: Remove alcohol
df_master = df_master[df_master['nutriscore_grade'] != 'not-applicable']

# Step 2: Create eco_packaging flag
eco_keywords = ['glass', 'paper', 'cardboard', 'recyclable-metals', 'paperboard']
df_master['eco_packaging'] = df_master['packaging_tags'].apply(
    lambda x: 1 if any(keyword in str(x).lower() for keyword in eco_keywords) else 0
)

# Step 3: Convert Nutri-Score to numbers
grade_map = {'a': 5, 'b': 4, 'c': 3, 'd': 2, 'e': 1}
df_master['nutriscore_num'] = df_master['nutriscore_grade'].map(grade_map)

# Step 4: Check the results!
print("Products per country after removing alcohol:")
print(df_master['country'].value_counts())
print("\nAverage Nutri-Score by eco_packaging:")
print(df_master.groupby('eco_packaging')['nutriscore_num'].mean())
print("\nAverage Nutri-Score by country AND eco_packaging:")
print(df_master.groupby(['country', 'eco_packaging'])['nutriscore_num'].mean())
```

```python
import itertools

# Drop empty values, flatten the lists, and get unique ones
all_categories = df_master['categories_tags'].dropna()
unique_categories = sorted(set(itertools.chain.from_iterable(all_categories)))

print("Unique categories found:")
for cat in unique_categories:
    print(f"  - {cat}")
```

```python
# Flatten the list of lists and count frequencies
from collections import Counter

all_categories_flat = []
for tags in df_master['categories_tags'].dropna():
    for tag in tags:
        all_categories_flat.append(tag)

category_counts = Counter(all_categories_flat)

# Show the top 10 most common categories
print("Most common categories:")
for category, count in category_counts.most_common(10):
    print(f"  {category}: {count} products")
```

```python
# Create a flattened version of the data (one row per category per product)
df_flattened = df_master[['product_name', 'categories_tags', 'nutriscore_num', 'eco_packaging']].copy()
df_flattened = df_flattened[df_flattened['categories_tags'].apply(lambda x: isinstance(x, list))]
df_flattened = df_flattened.explode('categories_tags')

# Calculate averages per category for eco and non-eco
category_summary = df_flattened.groupby(['categories_tags', 'eco_packaging']).agg(
    avg_score=('nutriscore_num', 'mean'),
    count=('product_name', 'count')
).reset_index()

# Pivot to have eco and non-eco side by side
pivot_table = category_summary.pivot(
    index='categories_tags',
    columns='eco_packaging',
    values=['avg_score', 'count']
).reset_index()

# Clean up column names
pivot_table.columns = ['category', 'avg_non_eco', 'avg_eco', 'count_non_eco', 'count_eco']

# Add total products and sample products
sample_products = df_flattened.groupby('categories_tags')['product_name'].apply(
    lambda x: list(x.head(3))
).reset_index()
sample_products.columns = ['category', 'sample_products']

# Merge everything
final_table = pivot_table.merge(sample_products, on='category')
final_table['total_products'] = final_table['count_non_eco'].fillna(0) + final_table['count_eco'].fillna(0)

# Filter to categories with at least 3 products and sort by total
final_table = final_table[final_table['total_products'] >= 3]
final_table = final_table.sort_values('total_products', ascending=False)

# Show the top 20 categories
final_table.head(20)
```

```python
# Filter for categories where Eco average is GREATER than Non-Eco
eco_better = final_table[final_table['avg_eco'] > final_table['avg_non_eco']].copy()

# Calculate the "Eco Advantage" (how much better is Eco?)
eco_better['advantage'] = eco_better['avg_eco'] - eco_better['avg_non_eco']

# Sort by biggest advantage first
eco_better = eco_better.sort_values('advantage', ascending=False)

# Show the results
print(f"🔥 Categories where Eco products are HEALTHIER than Non-Eco: {len(eco_better)}")
print("\nHere they are, sorted by biggest health advantage:\n")
print(eco_better[['category', 'avg_non_eco', 'avg_eco', 'advantage', 'total_products', 'sample_products']].head(10))
```

```python
# Categories where Eco is worse
eco_worse = final_table[final_table['avg_eco'] < final_table['avg_non_eco']]
print(f"❌ Categories where Eco products are WORSE than Non-Eco: {len(eco_worse)}")

# Categories where they are almost the same
eco_tie = final_table[final_table['avg_eco'] == final_table['avg_non_eco']]
print(f"⚖️ Categories where they are about equal: {len(eco_tie)}")
```

```python
#On average, eco-packaging is associated with worse health (2.33 vs 2.77).

#But that average hides a huge truth: 
#It is driven entirely by specific unhealthy categories that happen to use eco-packaging (e.g., sugary drinks in aluminium, chocolates in cardboard).

#The actionable recommendation:

#"If we want to sell 'healthy + sustainable', we should focus on categories where the overlap already exists—like breakfast cereals, crispbreads,
#and dairy. 
#For categories like beverages and snacks, we need to either improve the nutritional profile of the eco-packaged products, or clearly communicate 
#that eco-packaging is about the planet, not personal health."
```

```python
# 1. Create the organic flag (check if 'organic' appears in labels_tags)
df_master['organic'] = df_master['labels_tags'].apply(
    lambda x: 1 if isinstance(x, list) and any('organic' in tag for tag in x) else 0
)

# 2. Check the overall average Nutri-Score for Organic vs Non-Organic
print("Average Nutri-Score by Organic status:")
print(df_master.groupby('organic')['nutriscore_num'].mean())

# 3. Check the combination: Organic + Eco vs everything else
print("\nAverage Nutri-Score by combination of Organic and Eco:")
print(df_master.groupby(['organic', 'eco_packaging'])['nutriscore_num'].mean())
```
